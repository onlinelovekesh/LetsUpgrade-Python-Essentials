# Que. 1 Print the first armstrong number in range of 1042000 to 702648265 and exit the loop as soon you encounter the first armstrong number using while loop.


```python
first_no = 1042000
last_no = 702648265

for num in range(first_no, last_no + 1):
    
    order = len(str(num))
    
    sum = 0
    
    temp = num
    
    while temp > 0:
        
        digit = temp % 10
        sum += digit ** order 
        temp //= 10
    
    if num == sum:
        print("The first Armstrong number is : ", num)
        
        break;
        
        print(end = ' ')
```
