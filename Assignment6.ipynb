```python
def checkarmstrong(num):

    sum = 0

    temp = num
    while temp > 0:
        digit = temp % 10
        sum += digit ** 3
        temp //= 10

    if num == sum:
        return num
```

```python
checkarmstrong(371)
```

```python
lst = list(range(1,1000))
```

```python
def armstronggenerator(lst):
    for item in lst:
        sum = 0
        temp = item
        while temp > 0:
            digit = temp % 10
            sum += digit ** 3
            temp //= 10
        if item == sum:
            yield item
```

```python
print(list(armstronggenerator(lst)))
```
