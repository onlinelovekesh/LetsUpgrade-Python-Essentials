```python
# You all are Pilots, you want to land a plane safely, so, altitude required for landing a plane is 1000ft,
# if it is less than 1000ft then tell the pilot to "land the plane", or if it is more than that but less than 5000ft
# then tell the pilot to "come down to 1000ft", else if it is more than 5000ft#
# then tell the pilot to "Go around and try later".
```

```python
altitude = int(input("What is your current altitude? : "))

if altitude <= 1000:
    print("Land the plane")
    
elif altitude <= 5000:
    print("Come down to 1000ft")
    
else :
    print("Go around and try later")
```

```python
# Using for loop, Print all the prime numbers betweeen 1 and 200 using for loop and range function
```

```python
start = 1
end = 200
  
for i in range(start,end): 
    for j in range(2,i): 
        if(i % j == 0): 
            break
    else: 
        print(i)
```
