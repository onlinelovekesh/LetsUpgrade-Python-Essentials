# Sets and its Default Functions

```python
# set of integers


my_set = {1, 2, 3}
print(my_set)
```

```python
# set with different datatypes


my_set = {1.0, "Hello", (1, 2, 3)}
print(my_set)
```

```python
# add an element


my_set.add(2)
print(my_set)
```

```python
# add list and set


my_set.update([4, 5], {1, 6, 8})
print(my_set)
```

```python
# initializing my_set

my_set = {10,15,18,20,25}
print(my_set)
```

```python
# discard an element


my_set.discard(10)
print(my_set)
```

```python
# remove an element


my_set.remove(18)
print(my_set)
```

```python
# Negative indexing for accessing tuple elements


my_tuple = ('p', 'e', 'r', 'm', 'i', 't')

print(my_tuple[-1])
```
